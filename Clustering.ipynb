#KMEANS Clustering

selecting the optimal number of clusters is main and curicial part. The method which is very used very frequently is Elbow Method. In which we are actually varying the K from range of 1 to 10 and for each value of K, WCSS is calculated which is known as Within cluster sum of square.Once the ploting is done between K and WCSS, a graph with elbow is obtained. As the number of cluster tends to increase the value of WCSS will decrease. when K=1 , then WCSS have largest value.The K value corresponding to this point is the optimal K value or an optimal number of clusters.

K-Means Clustering is the most simplest and commonly used unsupervised learning.In Kmeans, we randomly give the K number of centroids in the data (mostly the value of K is found using elbow method as explained above ) and iterates these centroids until no change happens to the position of the centroid.
image.png

#Hierarchical Clustering

Hierarchical clustering is another unsupervised machine learning algorithm, which is used to group the unlabeled datasets into a cluster and also known as hierarchical cluster analysis or HCA.

In this algorithm, we develop the hierarchy of clusters in the form of a tree, and this tree-shaped structure is known as the dendrogram.

Sometimes the results of K-means clustering and hierarchical clustering may look similar, but they both differ depending on how they work. As there is no requirement to predetermine the number of clusters as we did in the K-Means algorithm. This method form the tree like structure. In this new clusters are formed using

Ecluidean Distance : This formula says the distance between two points (x1 1 , y1 1 ) and (x2 2 , y2 2 ) is d = √[(x2 – x1)2 + (y2 – y1)2].

selecting the optimal number of clusters is main and curicial part. The method which is very used very frequently is Elbow Method. In which we are actually varying the K from range of 1 to 10 and for each value of K, WCSS is calculated which is known as Within cluster sum of square.Once the ploting is done between K and WCSS, a graph with elbow is obtained. As the number of cluster tends to increase the value of WCSS will  decrease.
when K=1 , then WCSS have largest value.The K value corresponding to this point is the optimal K value or an optimal number of clusters.

K-Means Clustering is the most simplest and commonly used unsupervised learning.In Kmeans, we randomly give the K number of centroids in the data (mostly the value of K is found using elbow method as explained above ) and iterates these centroids until no change happens to the position of the centroid.  




```python
k = []
wcss = []
for i in range(2, 11):
  data = KMeans(n_clusters=i)
  data.fit(df_new.iloc[:,1:18])
  k.append(i)
  wcss.append(data.inertia_)

plt.plot(k, wcss,marker='o')
plt.xlabel('Values of K')
plt.ylabel('WCSS')
plt.title('Elbow Method To Find Number of Clusters')
plt.show()
```

```python
data = KMeans(n_clusters=4)
data.fit(df_new.iloc[:,1:18])
data.cluster_centers_
```


Hierarchical clustering is another unsupervised machine learning algorithm, which is used to group the unlabeled datasets into a cluster and also known as hierarchical cluster analysis or HCA.

In this algorithm, we develop the hierarchy of clusters in the form of a tree, and this tree-shaped structure is known as the dendrogram.

Sometimes the results of K-means clustering and hierarchical clustering may look similar, but they both differ depending on how they work. As there is no requirement to predetermine the number of clusters as we did in the K-Means algorithm.
This method form the tree like structure. In this new clusters are formed using

Ecluidean Distance : d = √[(x2 – x1)2 + (y2 – y1)2].


```python
import scipy.cluster.hierarchy as shc
import sklearn.cluster as sc
linked_data = shc.linkage(df_new.iloc[:,1:18].values,'ward',metric='euclidean')
plt.figure(figsize=(20,7))
shc.dendrogram(linked_data,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Dendrogram for the data')
plt.ylabel('Eculidean Distance')
plt.axhline(y=1.8, color='violet', linestyle='--')
plt.show()

```

Here, to do an internal valiadtion we have used silhouette score. The formula for silhouette score is  (b(i)-a(i)) / (max{b(i),a(i)})
Implemented a for loop for different number of clusters from 2 to 11

The silhouette value for Kmean clustering where k= 4 is 0.8750 and that of hierarchical clustering is 0.910.
Scores are high for both because it is matched with its own cluster but not matched with the other clusters.

```python
sil_coeff = []
for k in range(2, 11):
  data = KMeans(n_clusters=k)
  data.fit(df_new.iloc[:,1:18])
  sc = metrics.silhouette_score(df_new.iloc[:,1:18], data.labels_)
  sil_coeff.append(sc)
plt.plot(range(2, 11), sil_coeff,marker='o')
plt.xticks(range(2, 11))
plt.xlabel("No. of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```
